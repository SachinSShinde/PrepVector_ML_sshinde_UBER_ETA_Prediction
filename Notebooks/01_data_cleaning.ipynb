```python
# Required libraries
import numpy as np
import pandas as pd

# Scikit-Learn models, metrics, preprocessing
from sklearn.impute import SimpleImputer

#Filter and ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
def updateColumnName(df_train):
    df_train.rename(columns={'Weatherconditions': 'Weather_Conditions','multiple_deliveries': 'Multiple_Deliveries', 'City':'City_Type'}, inplace=True)

updateColumnName(df_train)
print(df_train.columns)
```

```python
# Extract time and convert to int
df_train['Time_taken(min)'] = df_train['Time_taken(min)'].apply(lambda x: int(x.split(' ')[1].strip()))
```

```python
# Extract Weather conditions
df_train['Weather_Conditions'] = df_train['Weather_Conditions'].apply(lambda x: x.split(' ')[1].strip())
```

```python
# Extract city code from Delivery person ID
df_train['City_Code'] = df_train['Delivery_person_ID'].str.split("RES", expand=True)[0]
```

```python
df_train[['ID', 'Time_taken(min)', 'Weather_Conditions', 'Delivery_person_ID', 'City_Code']].head()
```

```python
# Unique values of some categorical columns of df
columns = ['Weather_Conditions', 'Road_traffic_density',
       'Vehicle_condition', 'Type_of_order', 'Type_of_vehicle',
      'Multiple_Deliveries', 'Festival', 'City_Type']

for column in columns:
    unique_values = df_train[column].unique().tolist()
    print(column, ":", unique_values)
```

```python
df_train.drop(['ID','Delivery_person_ID'],axis=1,inplace=True)

df_train.shape[1]
```

```python
# Update datatypes
df_train['Delivery_person_Age'] = df_train['Delivery_person_Age'].astype('float64')
df_train['Delivery_person_Ratings'] = df_train['Delivery_person_Ratings'].astype('float64')
df_train['Multiple_Deliveries'] = df_train['Multiple_Deliveries'].astype('float64')
df_train['Order_Date']=pd.to_datetime(df_train['Order_Date'],format="%d-%m-%Y")
```

```python
# Converting String 'NaN' to np.nan
# As some of the values have extra space after NaN, will use regex
df_train.replace('NaN', float(np.nan), regex=True,inplace=True)
```

```python
df_train.isnull().sum().sort_values(ascending=False)
```

```python
# Updating null values as seen above
# Delivery_person_Ratings
df_train['Delivery_person_Ratings'].fillna(df_train['Delivery_person_Ratings'].median(), inplace=True)
df_train['Delivery_person_Age'].fillna(np.random.choice(df_train['Delivery_person_Age']), inplace=True)
df_train['Time_Orderd'] = df_train['Time_Orderd'].fillna(df_train['Time_Order_picked'])
df_train['Weather_Conditions'].fillna(np.random.choice(df_train['Weather_Conditions']), inplace=True)
```

```python
# Columns to update using SimpleImputer
update_cols = ['City_Type', 'Multiple_Deliveries', 'Road_traffic_density', 'Festival']
imputer_freq = SimpleImputer(missing_values=np.nan, strategy='most_frequent')

for col in update_cols:
    df_train[col] = imputer_freq.fit_transform(df_train[[col]]).ravel()  # Fix column reshaping issue
```

```python
df_train.isnull().sum()
```

```python
df_train.to_csv("../dataset/data_cleaned.csv")
```
